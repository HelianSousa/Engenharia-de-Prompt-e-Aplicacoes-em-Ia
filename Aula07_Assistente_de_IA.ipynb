# **Calculadora Simples**

```python
try:
    # Solicita ao usuário o primeiro número e converte para float
    num1 = float(input("Digite o primeiro número: "))
    # Solicita ao usuário o segundo número e converte para float
    num2 = float(input("Digite o segundo número: "))

    # Solicita a operação desejada (somar ou subtrair) e converte para minúsculas
    operacao = input("Escolha a operação (somar ou subtrair): ").lower()

    # Verifica a operação escolhida e realiza o cálculo correspondente
    if operacao == "somar":
        resultado = num1 + num2
        print(f"A soma de {num1} e {num2} é: {resultado}")
    elif operacao == "subtrair":
        resultado = num1 - num2
        print(f"A subtração de {num1} e {num2} é: {resultado}")
    else:
        # Exibe mensagem de erro se a operação for inválida
        print("Operação inválida. Por favor, escolha 'somar' ou 'subtrair'.")
except ValueError:
    # Captura erros caso o usuário não digite um número válido
    print("Entrada inválida. Por favor, digite apenas números para os valores.")
```
