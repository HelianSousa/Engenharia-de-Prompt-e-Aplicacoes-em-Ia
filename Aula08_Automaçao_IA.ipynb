```python
# Importa módulos necessários
import subprocess # Para executar comandos do sistema (instalar psutil)
import sys        # Para acessar variáveis e funções específicas do sistema (caminho do Python)

try:
    import psutil # Tenta importar a biblioteca psutil
except ImportError:
    # Se psutil não estiver instalado, imprime uma mensagem e tenta instalá-lo
    print("psutil não está instalado. Instalando...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "psutil"]) # Executa o comando pip install
    import psutil # Importa psutil novamente após a instalação

# Define o limite de temperatura para acionar o alerta de superaquecimento
SUPERHEATING_THRESHOLD = 32  # Temperatura limite em Celsius, ajustada para 32°C

print("Verificando a temperatura do sistema...")

# Obtém os dados de temperatura dos sensores do sistema usando psutil
temperatures = psutil.sensors_temperatures()

alert_triggered = False # Variável para controlar se algum alerta foi disparado

# Verifica se foram obtidos dados de temperatura
if temperatures:
    # Itera sobre cada tipo de sensor (ex: 'coretemp', 'acpitz')
    for name, entries in temperatures.items():
        print(f"--- {name.upper()} ---") # Imprime o nome do grupo de sensores
        # Itera sobre cada entrada de sensor dentro do grupo
        for entry in entries:
            # Obtém a temperatura atual, se disponível, caso contrário, define como None
            current = entry.current if hasattr(entry, 'current') else None
            # Usa o rótulo do sensor, se existir, caso contrário, usa o nome do grupo
            sensor_label = entry.label if entry.label else name

            print(f"  Sensor: {sensor_label}") # Imprime o rótulo do sensor
            # Verifica se a temperatura atual foi obtida
            if current is not None:
                print(f"    Atual: {current}°C") # Imprime a temperatura atual
                # Verifica se a temperatura atual excede o limite de superaquecimento
                if current > SUPERHEATING_THRESHOLD:
                    # Imprime um alerta de superaquecimento
                    print(f"    ALERTA: Superaquecimento detectado! {sensor_label} está em {current}°C, acima de {SUPERHEATING_THRESHOLD}°C.")
                    alert_triggered = True # Define que um alerta foi disparado
            else:
                print("    Atual: N/A") # Se a temperatura não foi obtida
            print("------------------")

    # Após verificar todos os sensores, se nenhum alerta foi disparado
    if not alert_triggered:
        print("Nenhum superaquecimento detectado. As temperaturas estão dentro dos limites normais.")

else:
    # Se não foi possível obter nenhum dado de temperatura dos sensores
    print("Não foi possível obter dados de temperatura dos sensores. Isso pode ocorrer em ambientes virtualizados ou onde os sensores não são expostos ao sistema operacional.")
    print("Para sistemas Linux, você pode tentar instalar 'lm-sensors' e executar 'sensors-detect' para habilitar a leitura dos sensores.")
```
